```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import math
```

```python
raw_data = pd.read_table('./traininingdata.txt', sep=';')
```

```python
raw_data.head()
```

```python
train_data = pd.read_csv('./trainingdata_processed.csv')
test_data = pd.read_csv('./testdata_processed.csv')
```

```python
train_data.head()
```

```python
df_inputs = train_data.drop('y', axis=1)
inputs = list(df_inputs.columns)
outputs = 'y'
```

```python
X_train = train_data[inputs].values.tolist()
X_test = test_data[inputs].values

y_train = train_data[outputs].values
y_test = test_data[outputs].values
```

```python
# do logistic regression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
# coefficients of logistic regression
lr.coef_
```

```python
# interception of logistic regression
lr.intercept_
```

```python
# feature importance
input_coef = {}
for i in range(len(inputs)):
    input_coef[inputs[i]] = lr.coef_[0][i] * np.std(X_train, 0)[i]
# sort by coef abs (feature importance)
input_coef_sorted = sorted(input_coef.items(), key=lambda x:abs(x[1]), reverse=True)
input_coef_sorted
```

```python
# visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
sns.set_theme(font_scale=0.5)
ax = sns.barplot(
    x=[i[1] for i in input_coef_sorted],
    y=[i[0] for i in input_coef_sorted],
    orient='h',
)
ax.bar_label(ax.containers[0])
```

```python
# predicted value 
y_hat_train = lr.predict(X_train)
y_hat_test = lr.predict(X_test)

#predicted probability 
prob_pred_train = lr.predict_proba(X_train)[:,1]
prob_pred_test = lr.predict_proba(X_test)[:,1]
```

### Confusion Matrix

```python
from sklearn.metrics import plot_confusion_matrix
```

```python
plot_confusion_matrix(lr, X_test, y_test)
```

### Precision / Recall / F-measure

```python
from sklearn.metrics import precision_recall_fscore_support
```

```python
precision_recall_fscore_support(y_test, y_hat_test, beta=1.0, labels=[1])
```

**For yes class:**
- **Precision**: 0.4722
- **Recall**: 0.4905
- **F-measure**: 0.4812

### Accuracy score

```python
from sklearn.metrics import accuracy_score
```

```python
print(
    f'Accuracy Score \
    \nTrain: {accuracy_score(y_train, y_hat_train)} \
    \nTest: {accuracy_score(y_test, y_hat_test)}'
)
```
