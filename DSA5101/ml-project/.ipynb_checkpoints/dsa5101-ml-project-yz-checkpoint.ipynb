```python
df = pd.read_csv('testdata.txt', sep=";", header=None)
df = df.rename(columns=df.iloc[0]).loc[1:]
convert_dict = {'age': int,
                'balance': int,
                'day': int,
                'duration': int,
                'campaign': int,
                'pdays': int,
                'previous': int,
                }
df = df.astype(convert_dict)

def data_pre_process(df):
    # drop columns
    df = df.drop(columns=['age', 'day', 'month', 'campaign', 'contact'])

#     # remove outliers for balance, duration, and previous columns using absolute z score
#     df = df[(np.abs(stats.zscore(df.balance)) < 3)]
#     df = df[(np.abs(stats.zscore(df.duration)) < 3)]
#     df = df[(np.abs(stats.zscore(df.previous)) < 3)]

    # binary columns
    df['default'].replace(['no', 'yes'],[0, 1], inplace=True)
    df['housing'].replace(['no', 'yes'],[0, 1], inplace=True)
    df['loan'].replace(['no', 'yes'],[0, 1], inplace=True)
    df['y'].replace(['no', 'yes'],[0, 1], inplace=True)

    # one hot encoding categorical columns
    df = pd.get_dummies(df, columns=['job', 'marital', 'education', 'poutcome'], \
                                prefix=['job', 'marital', 'edu', 'poutcome'])

#     # oversampling using SMOTE
#     x_train = df.drop(columns = ['y'])
#     y_train = df['y']
#     sm = SMOTE()
#     x_train_oversampled, y_train_oversampled = sm.fit_resample(x_train, y_train)
#     df_oversampled = pd.concat([x_train_oversampled, y_train_oversampled], axis=1)
    
    return df
```

```python
df = data_pre_process(df)
df.to_csv('testdata_processed.csv', index = False)
```
