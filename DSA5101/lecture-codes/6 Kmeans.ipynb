## Help on KMeans in module sklearn.cluster.k_means_ object. Some key parameters:

### n_clusters : int, optional, default: 8
>The number of clusters to form, as well as the number of
centroids to generate.
 
### max_iter : int, default: 300
>Maximum number of iterations of the k-means algorithm for a single run.
  
### n_init : int, default: 10
>Number of time the k-means algorithm will be run with different centroid seeds. The final results will be the **best output** of n_init consecutive runs in terms of inertia.
> Guess: evaluate how much distance do the clusters move, i.e. if the clusters travelled small distances than it is highly likely that we are closest to best solution.

### random_state : integer or numpy.RandomState, optional
>The generator used to initialize the centers. If an integer is
given, it fixes the seed. Defaults to the global numpy random
number generator.

### tol : float, default=1e-4
>  Relative tolerance with regards to  Euclidean/Frobenius norm of the difference
in the cluster centers of two consecutive iterations to declare
convergence.

## Attribute of the Kmeans clustering results
### inertia_ : float
> Sum of squared distances of samples to their closest cluster center.

## 3.1 Figure 1.1: Incorrect number of clusters (wrong K value): data X has three groups but we cluster it into two groups (K=2)

# Can we automatically decide the important paraemter k, i.e., n_clusters?

```python
SSE = []
for i in range(1, 15):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    #'k-means++' : selects initial cluster centers for k-mean clustering in a smart way to speed up convergence.
    # Number of time the k-means algorithm will be run with different centroid seeds. 
    # The final results will be the best output of n_init consecutive runs in terms of inertia.
    kmeans.fit(X)
    SSE.append(kmeans.inertia_)
    # Inertia =SSE, measures how well a dataset was clustered by K-Means. It is calculated by measuring the distance 
    # between each data point and its centroid, squaring this distance, and summing these squares.
plt.plot(range(1, 15), SSE)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.xlim([0, 15])
plt.show()
```

## It is amazing that we can automatically and accurately decide the number of clusters 3 using Elbow method!

However, it may not always work for other data sets or scenarios

## 3.2 Figure 2: Anisotropicly distributed data

```python
# Perform transformation and then clustering (with K=3)

transformation = [[ 0.60834549, -0.63667341], [-0.40887718, 0.85253229]]   # 2 by 2 matrix
X_aniso = np.dot(X, transformation)

# transformation here is to do matrix multiplication
# We have data X, which is a matrix [1500,2]
# We have a transformation matrix which is a matrix [2,2]
# x_aniso will be a matrix[1500,2]

# We group all the data into 3 clusters
y_pred_2 = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_aniso)

plt.scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred_2)
plt.title("Anisotropicly Distributed Blobs")
```

### 3.3 Figure 3: Different variance

```python
# Generate New 1500 (n_samples) Data, each cluster with different std (standard deviation)
# There is one cluster with big variance (std): 10 (yellow one)
# The remaining two clusters have small variances (std): 0.01 and 0.05
X_varied, y_varied = make_blobs(n_samples=n_samples,
                                cluster_std=[0.01, 0.05, 10],
                                random_state=random_state)
# X_varied stores data
# y_varied stores labels: 0, 1, 2. However, We do not use labels in this clusterng process 
# (we can evluate the accuracy if we need to do so)

# cluster_std: float or sequence of floats, optional (default=1.0)
# The standard deviation of the clusters.

# Different clusters have different variance. 
# Specificially, cluster 2 has largest variance

y_pred_3 = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_varied)
plt.scatter(X_varied[:, 0], X_varied[:, 1], c=y_pred_3)
plt.title("Unequal Variance")
# It seems yellow color cluster has largest gradient
```

### 3.4 Figure 4: Unevenly sized blobs

```python
# Generate New Data again - this time unevenly sized blob
X_filtered = np.vstack((X[y == 0][:1000], X[y == 1][:200], X[y == 2][:20]))

# Cluster 1 has 1000 points, Cluster 2 has 200 points, cluster 3 has 20 points

# Performing KMeans clustering
y_pred_4 = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_filtered)

plt.scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_pred_4)
plt.title("Unevenly Sized Blobs")
```

### subplot(nrows, ncols, plot_number)
    
> Where *nrows* and *ncols* are used to notionally split the figure into ``nrows * ncols`` sub-axes, 

> *plot_number* is used to identify
the particular subplot that this function is to create within the notional
grid. *plot_number* starts at 1, increments across rows first and has a
maximum of ``nrows * ncols``. It basically finishes the 1st row and then continues with 2rd row, ....
    
#### In the case when *nrows*, *ncols* and *plot_number* are all less than 10, a convenience exists, such that the a 3 digit number (e.g. 211) can be given instead, where the hundreds represent *nrows*, the tens represent *ncols* and the units represent *plot_number*. For instance
    
      subplot(221)
      
     

### produces a subaxes in a figure which represents the top left plot (i.e. the first) in a 2 row by 2 column notional grid (no grid actually exists, but conceptually this is how the returned subplot has been positioned).


Grid|Col1|Col2
-|-|-
Row1|221|222
Row2|223|224


```python
# set the spacing between subplots
# We can use the plt.subplots_adjust() method to change the space between Matplotlib subplots. 
#The parameters wspace and hspace specify the space reserved between Matplotlib subplots. 
#They are the fractions of axis width and height, respectively. And the parameters left, right, top and bottom parameters 
#specify four sides of the subplots’ positions. They are the fractions of the width and height of the figure.
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
plt.subplot(221)
# Grid with 2 Rows, 2 Column, we will draw the results in the first figure
plt.scatter(X[:, 0], X[:, 1], c=y_pred_1)   # All rows and two columns (we draw 1500 2D data)
plt.title("Incorrect Number of Blobs")

plt.subplot(222)
# Grid with 2 Rows, 2 Column, the second figure
plt.scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred_2)
plt.title("Anisotropicly Distributed Blobs")

plt.subplot(223)
# Grid with 2 Rows, 2 Column, the third figure
plt.scatter(X_varied[:, 0], X_varied[:, 1], c=y_pred_3)
plt.title("Unequal Variance")

plt.subplot(224)
# Grid with 2 Rows, 2 Column, the fourth figure
plt.scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_pred_4)
plt.title("Unevenly Sized Blobs")

#plt.show()
```
