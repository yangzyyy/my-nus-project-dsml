```python
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt

from wordcloud import WordCloud
```

```python
df = pd.read_csv('TweetsCOV19.tsv', sep='\t', header=None, on_bad_lines='skip')
```

```python
df.columns = ['TweetID', 'Username', 'Timestamp', 'N_Followers', 'N_Friends', 'N_Retweets', 'N_Favorites',
             'Entities', 'Sentiment', 'Mentions', 'Hashtags', 'URLs']
```

```python
len(df)
```

```python
df.isnull().sum()
```

```python
df = df[df.Entities != 'null;']
```

```python
len(df)
```

```python
df['Original_Entities'] = [[entity.split(':')[0] for entity in entity_group.split(';')[:-1]] 
                           for entity_group in df['Entities']]
df['Annotated_Entities'] = [[entity.split(':')[1] for entity in entity_group.split(';')[:-1]] 
                            for entity_group in df['Entities']]
df['Timestamp'] = pd.to_datetime(df['Timestamp']).dt.strftime("%Y-%m-%d")
df['Positive_Sentiment'] = [int(x[0]) for x in df['Sentiment']]
df['Negative_Sentiment'] = [int(x[-2:]) for x in df['Sentiment']]
df['Overall_Sentiment'] = df['Positive_Sentiment'] + df['Negative_Sentiment']
df = df.drop(['Username', 'Entities', 'URLs', 'Sentiment'], axis=1)
```

```python
df.to_csv('processed_data.csv', index=False)
```

```python
df.dtypes
```

```python
df.head()
```

```python
sns.set_theme()
fig, ax = plt.subplots(1, 2, figsize=(16, 8), sharey=True)

sns.histplot(
    x='Positive_Sentiment',
    data=df,
    ax=ax[0],
    stat='percent'
)

sns.histplot(
    x='Negative_Sentiment',
    data=df,
    ax=ax[1],
    stat='percent'
)

plt.show()
```

```python
sns.set_theme()
fig, ax = plt.subplots(1, 2, figsize=(16, 8), sharey=True)

sns.scatterplot(
    x='Positive_Sentiment',
    y='N_Retweets',
    data=df[df['N_Retweets'] != 0],
    ax=ax[0]
)

sns.scatterplot(
    x='Negative_Sentiment',
    y='N_Retweets',
    data=df[df['N_Retweets'] != 0],
    ax=ax[1]
)

plt.show()
```

```python
df.describe()
```

```python
df_retweet_0 = df[df['N_Retweets'] == 0]
df_retweet_not_0 = df[df['N_Retweets'] != 0]

fig, ax = plt.subplots(1, 2, figsize=(16, 8))

sns.scatterplot(
    x='Overall_Sentiment',
    y='N_Retweets',
    data=df_retweet_not_0,
    ax=ax[0]
)

sns.histplot(
    x='Overall_Sentiment',
    data=df_retweet_0,
    ax=ax[1],
    stat='percent'
)
plt.show()
```
