## Question 2 (4 points)

Perform linear regression using all the input variables on this dataset to predict the compressive strength. Carefully evaluate your regression model using appropriate metrics and datasets. 

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression;
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
````

Split dataset into training set and testing set.

````python
data_train, data_test = train_test_split(data, test_size=0.1)
````

Separate X values(input) and y values(output).

````python
x_train = data_train.drop('Concrete_Compressive_Strength',axis= 1)
x_test = data_test.drop('Concrete_Compressive_Strength',axis= 1)

y_train = data_train['Concrete_Compressive_Strength']
y_test = data_test['Concrete_Compressive_Strength']
````

Apply simple linear regression on all input variables, and predict output using the trained model called regressor.

````python
regressor = LinearRegression()
regressor.fit(
    X=x_train,
    y=y_train,
)
y_hat_train = regressor.predict(x_train)
y_hat_test = regressor.predict(x_test)
````

Evaluate regressor model using R^2 score and RMSE.

````python
def rmse_scaled(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    scale = np.sqrt(np.mean(y_true**2))
    return rmse/scale

print('r2 score for data_train : ', r2_score(y_train, y_hat_train))
print('r2 score for data_test : ', r2_score(y_test, y_hat_test))
print('Scaled RMSE for data_train : ', rmse_scaled(y_train, y_hat_train))
print('Scaled RMSE for data_test : ', rmse_scaled(y_test, y_hat_test))
````

## Question 3 (4 points)

Now, instead of linear regression, apply kernel ridge regression (documentation [here](https://scikit-learn.org/stable/modules/generated/sklearn.kernel_ridge.KernelRidge.html)) with 3 kernels:
1. linear kernel
2. polynomial kernel with degree 3
3. rbf kernel

Quantify the performance improvement (if any) with respect to the linear case, and also the other kernels. What do you observe? Which would you choose?

*Hint: For the RBF kernel, should you do something to the inputs first? Why?*

---

Note: we have not gone through cross validation and hyper-parameter tuning yet, so you are not expected to do this here -- although you are free to do it if you know how. 

````python
from sklearn.kernel_ridge import KernelRidge
from sklearn import preprocessing
````

1. Linear Kernel

````python
krr_linear = KernelRidge()
krr_linear.fit(X=x_train,y=y_train,)
y_hat_train_krr_l = krr_linear.predict(x_train)
y_hat_test_krr_l = krr_linear.predict(x_test)

print('r2 score for data_train : ', r2_score(y_train, y_hat_train_krr_l))
print('r2 score for data_test : ', r2_score(y_test, y_hat_test_krr_l))
print('Scaled RMSE for data_train : ', rmse_scaled(y_train, y_hat_train_krr_l))
print('Scaled RMSE for data_test : ', rmse_scaled(y_test, y_hat_test_krr_l))
````

2. Polynomial Kernel with degree 3

````python
krr_poly = KernelRidge(kernel='polynomial',degree=3)
krr_poly.fit(X=x_train,y=y_train,)
y_hat_train_krr_p = krr_poly.predict(x_train)
y_hat_test_krr_p = krr_poly.predict(x_test)

print('r2 score for data_train : ', r2_score(y_train, y_hat_train_krr_p))
print('r2 score for data_test : ', r2_score(y_test, y_hat_test_krr_p))
print('Scaled RMSE for data_train : ', rmse_scaled(y_train, y_hat_train_krr_p))
print('Scaled RMSE for data_test : ', rmse_scaled(y_test, y_hat_test_krr_p))
````

3. RBF Kernel
<br>
Before training the RBF kernel, standard scaler is used to standardize all features for better outcome.

````python
scaler = preprocessing.StandardScaler()
x_scaled_train = scaler.fit_transform(x_train)
x_scaled_test = scaler.fit_transform(x_test)

krr_rbf = KernelRidge(kernel="rbf", gamma=0.1)
krr_rbf.fit(X=x_scaled_train,y=y_train,)
y_hat_train_krr_rbf = krr_rbf.predict(x_scaled_train)
y_hat_test_krr_rbf = krr_rbf.predict(x_scaled_test)

print('r2 score for data_train : ', r2_score(y_train, y_hat_train_krr_rbf))
print('r2 score for data_test : ', r2_score(y_test, y_hat_test_krr_rbf))
print('Scaled RMSE for data_train : ', rmse_scaled(y_train, y_hat_train_krr_rbf))
print('Scaled RMSE for data_test : ', rmse_scaled(y_test, y_hat_test_krr_rbf))
````

````python
r2_score_train = [r2_score(y_train, y_hat_train_krr_l), r2_score(y_train, y_hat_train_krr_p), r2_score(y_train, y_hat_train_krr_rbf)]
r2_score_test = [r2_score(y_test, y_hat_test_krr_l), r2_score(y_test, y_hat_test_krr_p), r2_score(y_test, y_hat_test_krr_rbf)]
scaled_rmse_train = [rmse_scaled(y_train, y_hat_train_krr_l), rmse_scaled(y_train, y_hat_train_krr_p), rmse_scaled(y_train, y_hat_train_krr_rbf)]
scaled_rmse_test = [rmse_scaled(y_test, y_hat_test_krr_l), rmse_scaled(y_test, y_hat_test_krr_p), rmse_scaled(y_test, y_hat_test_krr_rbf)]

score = [r2_score_train, r2_score_test, scaled_rmse_train, scaled_rmse_test]
score = [[np.round(float(i), 2) for i in nested] for nested in score]

df_compare = pd.DataFrame(score, columns=['Linear','Polynomial','RBF'], \
                          index = ['r2_score_train', 'r2_score_test', 'scaled_rmse_train', 'scaled_rmse_test'])
df_compare = df_compare.melt(ignore_index=False, value_vars = ['Linear', 'Polynomial', 'RBF'], var_name='Kernel', value_name='Values')

sns.set(font_scale = 0.7)
ax = sns.barplot(x=df_compare.index, y='Values', data=df_compare, hue='Kernel')
for i in ax.containers:
    ax.bar_label(i,)
````

Compared to the linear kernel model, polynomial kernel and RBF kernel train better models and predict more accurately on both training and testing data sets. Polynomial kernel model performs the best on the trainig set, but has almost the same accuracy as RBF kernel on the testing set.

I would choose RBF kernel because it trains the model that performs consistently on training and testing sets.
