```python
# demonstration of using numpyvalues = [[23.1,45.2, 94.4, 34.2],[10.0,12.3,20,15]]
import numpy as np
values = [[23.1,45.2, 94.4, 34.2],[10.0,12.3,20,15]]  #generic array
np_array = np.array (values)    #numpy array 
print ('The shape of the array: ',np.shape(np_array))
print (np_array)
print ('Scalar multiplication and addition')
print (np_array * 9 /5 +32)
```

```python
# Using index in numpy
print ('The numpy array')
print(np_array)
print('The last row: ')
print (np_array[-1,:])
print('The block with 1--2 rows and 2--3 column')
print (np_array[0:2,1:3])
```

```python
# numpy supports complex-valued number and operation
a = np.array([1+2j, 3+4j, 5+6j])
print('The vector a: ', a)
print('Its real part: ', a)
print('Its imaginary part: ', np.real(a))
print('Its magnitude: ', np.abs(a))
print('Its angle: ', np.angle(a))  #in radian
```

```python
# Using linear algebra operations in numpy
A = np.reshape ([1,2,3,4], [2,2])
B = np.reshape ([5,6,7,8], [2,2,])
print (A, ' : The matrix A')
print (B, ' : The matrix B')
print (A * B, ' : The elementwise multiplication')
print (A @ B, ' : The matrix multiplication')
```

```python
A = np.array([[1,3,4],[4,6,7],[0,10,2]])
b = np.array([1,3,5])
# Solve the linear system Ax=b
x = np.linalg.solve(A,b)
np.set_printoptions(precision=2)
print('The LHS matrix A \n',A)
print('The RHS vector b\n',b)
print('The solution x\n',x)
print('The residual Ax-b is: ', A@x-b)
```

```python
# Converting a color image to a gray-scale image

im = np.array(image.open('pepper.png'))
def rgb2gray (rgb):
    return np.dot(rgb[:,:,:3], [0.299, 0.587, 0.144])
im2gray = rgb2gray (im)
print(im2gray)
plt.subplot(1,2,1); plt.imshow(im);
plt.subplot(1,2,2); plt.imshow(im2gray, cmap = 'gray');
```

```python
# Image cropping
im = np.array(image.open('pepper.png'))
im_crop = im [20:-70, 15:-50,: ]
plt.subplot(1,2,1); plt.imshow(im);
plt.subplot(1,2,2); plt.imshow(im_crop);
```

```python
# Image rotating

im = np.array(image.open('pepper.png'))
im_T = np.transpose(im, (1,0,2))
im_rot90 = im_T[::-1,:,:]
plt.subplot(1,2,1); plt.imshow(im);
plt.subplot(1,2,2); plt.imshow(im_rot90);
```

```python
# Spectral analysis for 1D signal

x = np.array([1.5, 3.1, 2.4, -0.5, -4])
x_hat = np.fft.fft(x)
x_hat2 = np.fft.fftshift (x_hat)
x_rec = np.fft.ifft(x_hat)
np.set_printoptions(formatter={'complex_kind': '{:.2f}'.format})
print('Signal: ',x)
print('FFT: ', x_hat)
print('FFT with fftshift: ', x_hat2)
print('iFFT: ', x_rec)
```

```python
# Using fftshift for re-align frequency
grid = np.linspace(1,20,201)
x = np.sin(grid)
fx_1 = np.fft.fft(x)
fx_2 = np.fft.fftshift (fx_1)
fig = plt.figure()
fig.set_figheight(2)
fig.set_figwidth(12)
plt.subplot(1,3,1); plt.plot(x);
plt.subplot(1,3,2); plt.plot(abs(fx_1));
plt.subplot(1,3,3); plt.plot(abs(fx_2));
```

```python
# Spectral analysis for 2D images
im = np.array(image.open('cameraman.png'))
print('The dimension of image: ', np.shape(im))
im_hat = np.fft.fftshift(np.fft.fft2(im))
print('The dimension of 2DFFT: ', np.shape(im_hat))
im_rec = np.fft.ifft2(np.fft.ifftshift(im_hat))
plt.subplot(1,4,1); plt.imshow(im, cmap = 'gray')
plt.subplot(1,4,2); plt.imshow(np.real(im_hat), cmap = 'gray')
plt.subplot(1,4,3); plt.imshow(np.imag(im_hat), cmap = 'gray')
plt.subplot(1,4,4); plt.imshow(np.real(im_rec), cmap = 'gray')
```

```python
# 2D Image convolution by a high-pass filter

from scipy.signal import convolve2d
im = np.array(image.open('cameraman.png'))
K = 1/4 * np.array([[-1,0,-1],[0,4,0], [-1,0,-1]])
print(K, ': 2D filter')
plt.subplot(1,2,1); plt.imshow(im, cmap = 'gray');
plt.subplot(1,2,2); plt.imshow(convolve2d(im,K,'same'), cmap= 'gray');
```
